```python
from IPython.display import Image
Image(filename='img/img-api-salic.png',width=800, height=400)
```

```python
import requests
```

```python
#Processo de paginação:
#Limite: Quantidade de projetos retornados
#Offset: A partir de qual projeto você uqer visualizar

response = requests.get("http://api.salic.cultura.gov.br/v1/projetos/?limit=100&offset=0")
response.json()
```

```python
#Processo automatizado

result = pd.DataFrame()

#Processeo de leitura das páginas
for i in range(0, 5000, 100):
    response = requests.get("http://api.salic.cultura.gov.br/v1/projetos/?limit=100&offset="+str(i))
    response = response.json()['_embedded']['projetos']
    
    #Processo de leitura dos projetos do json
    for j in response:
        j.pop('_links', None)
        temp = pd.DataFrame([j])
        result = result.append(temp, ignore_index=True)
```

```python
result
```

```python
result['data_inicio'].min()
```

```python
result['data_inicio'].max()
```

```python
result = result[['nome', 'segmento', 'area', 'data_inicio', 'UF', 'municipio','valor_captado', 'valor_projeto', 'valor_proposta', 'valor_aprovado']]
result.head()
```

```python
result = result.loc[result['data_inicio'] > '2000-01-01']
```
